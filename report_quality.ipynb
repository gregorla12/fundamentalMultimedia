```python
import cv2 as cv2
import matplotlib.pylab as plt
```

```python
img1= cv2.imread('test.png', cv2.COLOR_BGR2RGB)# orgininal image
img1_rec= cv2.imread('rec.png', cv2.COLOR_BGR2RGB)# restored image

cv2.PSNR(img1, img1_rec)
```

```python
plt.figure(figsize=(17, 17))
plt.subplot("121"), plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)), plt.title("original")
plt.subplot("122"), plt.imshow(cv2.cvtColor(img1_rec, cv2.COLOR_BGR2RGB)), plt.title("rec")
```
